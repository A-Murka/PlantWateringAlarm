```python
data=np.array(loadtxt('version-test.csv', delimiter=','))
```

```python
v1moist = data[:,0]/471*100
v1moist = v1moist - min(v1moist)
v2moist = data[:,3]/762*100
v2moist = v2moist - min(v2moist)

tempic=data[:,1]
temp=data[:,4]/10
figure(figsize=(10, 10))
plot( temp, v1moist, label="v.2.5 (old)", color="red")
plot( temp, v2moist, label="v2.7.1 (new)", color="green")
title("Moisture reading drift with temperature")
xlabel("Temperature, C")
ylabel("Relative moisture reading, %")
legend()
```

```python
drift=v2moist[-1]-v2moist[0]
dt=temp[-1]-temp[0]
drift/dt
```
